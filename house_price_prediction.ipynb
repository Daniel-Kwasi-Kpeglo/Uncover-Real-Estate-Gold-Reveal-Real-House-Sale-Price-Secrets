# Background

## Understanding House Price Prediction
- **Objective**: Predict the selling prices of houses based on various features like location, size, age, etc.
- **Importance**: Accurate predictions help buyers and sellers make informed decisions, and assist real estate agents and economists in market analysis.

## Role of Machine Learning in this Project
- **Machine Learning in Real Estate**: Utilizes historical data to forecast future prices, making predictions more accurate and reliable.
- **Random Forest Algorithm**:
  - **Type**: A popular, power machine learning model.
  - **Working Principle**: Builds multiple decision trees and merges them to get a more accurate and stable prediction.
  - **Advantages**: Handles a large number of features, works well with non-linear data, and provides feature importance.

## Incorporating SHAP for Explainability
- **SHAP (SHapley Additive exPlanations)**:
  - **Purpose**: Explains the output of machine learning models.
  - **Methodology**: Uses game theory approach to assign each feature an importance value (shap value) for a particular prediction.
- **Benefits of SHAP in the Project**:
  - **Transparency**: Helps understand why the model made a specific prediction.
  - **Insightful**: Identifies which features are most influential in determining house prices.
  - **Trustworthy**: Makes the model's decisions interpretable to humans, increasing trust and usability.

### Dataset description

This dataset contains house sale prices for King County, which includes Seattle. It includes homes sold between May 2014 and May 2015. <br>

It's a great dataset for evaluating simple regression models.

- **Id**: Unique ID for each home sold
- **Date**: Date of the home sale
- **Price**: Price of each home sold
- **Bedrooms**: Number of bedrooms
- **Bathrooms**: Number of bathrooms, where .5 accounts for a room with a toilet but no shower
- **Sqft_living**: Square footage of the apartments interior living space
- **Sqft_lot**: Square footage of the land space
- **Floors**: Number of floors
- **Waterfront**: A dummy variable for whether the apartment was overlooking the waterfront or not
- **View**: An index from 0 to 4 of how good the view of the property was
- **Condition**: An index from 1 to 5 on the condition of the apartment
- **Grade**: An index from 1 to 13, where 1-3 falls short of building construction and design, 7 has an average level of construction and design, and 11-13 have a high quality level of construction and design
- **Sqft_above**: The square footage of the interior housing space that is above ground level
- **Sqft_basement**: The square footage of the interior housing space that is below ground level
- **Yr_built**: The year the house was initially built
- **Yr_renovated**: The year of the house's last renovation
- **Zipcode**: What zipcode area the house is in
- **Lat**: Latitude
- **Long**: Longitude
- **Sqft_living15**: The square footage of interior housing living space for the nearest 15 neighbors
- **Sqft_lot15**: The square footage of the land lots of the nearest 15 neighbors


### Prepare dataset


#### Import data


````python
await skillsnetwork.download_dataset('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMSkillsNetwork-GPXX0GJDEN/kc_house_data.csv')
df = pd.read_csv('kc_house_data.csv')
````

#### Take a look at dataset
Using function `sample(x)` to randomly show x numbers of samples


````python
df.sample(5)
````

#### Drop the column **Id** in the dataset
As columns **Id**, **date**  will not be used in the analysis, we drop it using function `drop()`.
- `axis=1` indicates that it is dropping column instead of a row.
- `inplace=True` indicates that we modify the dataframe itself instead of creating a new one.


````python
df.drop('id', axis=1, inplace=True)
df.drop('date', axis=1, inplace=True)
````

We can inspect the types of feature columns using the following code:


````python
df.info()
````

Plot house price distribution, we see house 


````python
plt.figure(figsize=(6, 4))
sns.distplot(df['price'], color='b', bins=100, hist_kws={'alpha': 0.4});
````

### Data Spliting
In order to conduct the supervised learning, we need to split data to get X_train, X_test, y_train, y_test


````python
X = df.drop(columns = ['price'])
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
````

### Modeling


#### Set parameters for the model


````python
param_randomf = {
    'n_estimators' : [100, 200],
    'max_depth': [10, 20]
}
````

Using `GridSearchCV` function from scikit-learn for hyperparameter tuning. Let's break down the code here. <br>
- `estimator=RandomForestRegressor()`: Here, RandomForestRegressor is set as the estimator for the grid search. This means that the grid search will be tuning a Random Forest regression model.
- `param_grid=param_randomf`: The param_grid specifies the parameters, which will use the parameters set in `param_randomf`.
- `cv=5`: This indicates that 5-fold cross-validation is used. In cross-validation, the dataset is divided into 'k' parts (in this case, 5 parts). The model is trained on 'k-1' parts and validated on the remaining part. <br><br>
It might take around 3~5 minutes. Let's take a break.


````python
grid_search = GridSearchCV(estimator=RandomForestRegressor(), param_grid=param_randomf, cv=5)
grid_search.fit(X_train, y_train)
````

After fits the Random Forest model to the training data, trying out all the different combinations of parameter values specified in `param_randomf`, we can get a best model.


````python
best_model = grid_search.best_estimator_
````

We can take a look at the best parameters


````python
best_params = grid_search.best_estimator_
best_params
````

Using best model to test on the X_test dataset and get the results of prediction.


````python
y_pred = best_model.predict(X_test)
````

Using metric R² to present the performance of the model.
- `r2_score`: This function comes from scikit-learn. It is used to calculate the R² score, which is a statistical measure of how close the data are to the fitted regression line. In simpler terms, it measures how well the regression predictions approximate the real data points.
- The R² score is a number between 0 and 1, where a score closer to 1 indicates that the model's predictions are very close to the actual values.


````python
r2 = r2_score(y_test, y_pred)
r2
````

From the observed results, given the current parameter settings, it is evident that the R² score is approximately 0.859. This score suggests a strong performance of the regression model in the given task.

# Using SHAP to explain the model


Create a subset of training data (100 samples). Computing SHAP values for the features in `X_sample`, as if we use whole dataset, it will take significant time for shap to process. Of course you can set more data or even whole traing data.
- Using `shap.Explainer` to explain the `best model` we get previously on `X_sample`
- Using  `shap_values` variable to hold the SHAP values computed by the explainer. These values can be used to understand how each feature in the data contributes to each prediction made by your model.


````python
X_sample = X_train.sample(100)
explainer = shap.Explainer(best_model, X_sample)
shap_values = explainer(X_sample)
````

Using beeswarm plot to show shap_values.


````python
shap.plots.beeswarm(shap_values)
````

The bee-swarm plot of the SHAP values offers valuable insights into the factors that influence house selling prices. 
- The vertical axis placement indicates the magnitude of effect a feature has on the price. 
- The horizontal axis represents the direction of the effect: values to the left decrease the price, while values to the right increase it. 
- Each data point is represented as a dot or bee along the horizontal axis. Using colour to indicating its value within the specified feature (red indicates high values, while blue represents low values).


Key observations include:
- Location is the most critical determinant of house prices.
- The living area size, especially the square footage, is a significant predictor of price, with larger areas boosting the predicted value.
- The property's grade and age are also influential, with older properties generally reducing the predicted selling price.
- Additional factors such as the number of bedrooms and basement size, along with seven other attributes, have a comparatively minor effect on pricing.


We also can see how each feature's influence to house price prediction using `force_plot` in SHAP.


````python
shap.initjs()
shap.plots.force(shap_values)
````

Let's examine an example using the output plot. Once the plot is generated, set the y-axis to represent the effects of sqft_living, and the x-axis to display sqft_living values. This setup allows us to observe the specific points at which sqft_living has either a positive or negative impact on the model.

- Notably, a sqft_living value of approximately 2000 serves as a pivotal threshold. Values greater than 2000 sqft_living tend to have a positive influence, while those less than 2000 exhibit a negative impact.


Take away points both house sellers or buyers

- **Location is Key**: The location of a property is one of the most significant determinants of its value. Buyers should prioritize location based on their lifestyle needs and future resale value, while sellers should highlight the benefits of their property's location.

- **Size Matters, Especially Living Space**: The square footage of the living area is a major price influencer. Buyers should consider their space requirements carefully, and sellers should emphasize the spaciousness and layout of living areas.

- **Age and Condition of the Property**: Newer properties often fetch higher prices. Sellers with older properties should consider upgrades or renovations to improve value. Buyers should assess the age and condition of the property in terms of potential future maintenance and upgrade costs.

- **Property Grade and Features**: The overall grade of the property, which includes construction quality and design, impacts its value. Buyers should be mindful of these aspects for long-term satisfaction, while sellers should highlight these features in their marketing.
